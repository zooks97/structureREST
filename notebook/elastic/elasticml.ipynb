```python
import math
def soap_kernel(x1, x2):
    k11 = np.linalg.norm(x1)
    k22 = np.linalg.norm(x2)
    k12 = min(np.dot(x1, x2)/(k11*k22), 1.0)
    dist = np.sqrt(2-2*k12)
    print(dist)
    return dist

frame["K_VRH"] = frame["elasticity"].apply(lambda x: x["K_VRH"])
y = np.array(frame["K_VRH"])
```
