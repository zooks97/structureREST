```python
# -*- coding: utf-8 -*-
import requests
from pymatgen.io.cif import CifParser
from pymatgen import Structure
from glob import glob
import json
from tabulate import tabulate
```

```python
display(sorted(glob('../../data/*cif')))
```

```python
cifs = glob('../../data/Li*cif')
structures = [CifParser(cif).get_structures(primitive=False)[0].as_dict() for cif in cifs]
payload = {'structures': json.dumps(structures)}
```

```python
endpoints = {'distances': ['matminer', 'pymatgen'],
             'comparisons': ['matminer', 'pymatgen'],
             'fingerprints': ['matminer', 'stidy']}
for endpoint, functions in endpoints.items():
    for function in functions:
        print(endpoint, function)
        r = requests.get('http://127.0.0.1:5000/v1/{}/{}/'.format(endpoint, function), params=payload)
        results = json.loads(r.text)
        print(results)
```
